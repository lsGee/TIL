## 1. 함수

**1. 함수 정의 시 (가변 아규먼트 관련)**
  * 파라미터에 * 붙이면, 붙은 대상을 "튜플"로 패킹하여 받음
    </br>→ 가변인수 (= 변수 여러 개 가능)
  * \*\*의 의미는 키워드아규먼트를 여러 개(가변) 받겠다는 의미
    </br>→ 함수 선언 시 가장 뒤에 와야 하는 파라미터! 
  * (언패킹 시) \*args는 튜플, \**kwargs는 딕셔너리
  * 함수 정의 시 파라미터 순서 : 위치매개변수(기본값X) → \*args → \**kwargs
    </br>※ 기본값 있는 매개변수는 \*args 뒤에 와도 됨(but,, \*\*kwargs보다는 앞)
  * 기본값 지정된 파라미터는 기본값 없는 파라미터의 뒤에 나와야 함
</br></br>

**2. 함수 호출 시 (가변 아규먼트 관련)**
  * 아규먼트에 *붙이면, * 붙은 대상을 언패킹하여 전달
  * 언패킹할 매개변수는 튜플 외 시퀀스도 가능
  * 아규먼트 중 키워드아규먼트는 포지션아규먼트 뒤에 나와야 함
  * 키워드아규먼트는 꼭 있어야 하는 건 아님(but 함수 내 해당 키워드인수 사용 시, 없이 호출하면 에러날 수 있음)


</br>※ 추가참고: https://brunch.co.kr/@princox/180

```
def test(n, *args, alive=True,**kwargs):
    print('n:', n)
    for i in args:
        print(i, end=' ')
    print('\nalive =', alive)
    if 'gender' in kwargs.keys():
        print('당신은', kwargs['gender'])
    else: print('성별(gender)을 입력하라구')
    print()

test('상상', '아아', 1, 2, 3333, gender='여자')

##### 아래는 리스트 매개변수 언패킹해 전달 #####
t = [10, 20, '가나다']
# t = 'abcd'
test('상상', *t, alive=False, gender='female')
```

### 추가) 언패킹 (vs. 패킹)

* 언패킹 : 콜렉션(시퀀스 및 이터러블 객체) -> 개별 값(각 값들을 서로다른 변수들에 저장)
* 패킹 : 개별 값 -> 콜렉션(튜플)
</br>

```
############ 언패킹 ############
# 튜플 s를 s1과 s2 변수로 언패킹
s = 1, 2
s1, s2 = s  ## 언패킹!
print("변수에 넣지 않고 바로 언패킹", *s)   # 변수에 넣지 않고 바로 언패킹

# 리스트 l을 l1, l2 , l3변수로 언패킹
l = [1, 2, 3]
l1, l2, l3 = l 

print(s1, s2)
print(l1, l2, l3)

# range 객체를 언패킹
# 주의: range 객체 길이와 변수 개수를 맞춰줘야 함..!
r1,r2 = range(2)
print(r1,r2)

############# 패킹 #############
# s1과 s2를 패킹
rs = s1, s2
print(type(rs), rs)

# l1, l2, l3를 패킹
rl = l1, l2, l3
print(type(rl), rl)
```

```
## 언패킹 응용 (1)
## 리스트를 한번에 출력 후 요소 사이에 sep 적용하고 싶을 때
## sep 적용을 위해 for 문 사용 필요 없음

li = [1,2,3,4]
print(li)
print(*li, sep='-')
```

```
## 언패킹 응용 (2)
## 언더바(_) 이용 시 특정 자리 값만 언패킹할 수 있음
li = [1,2,3,4,5]

# _
a,_,_,_,b = li
print(a, b)

# *_
c,*_,d = li
print(c,d)
```

### 2-2. 딕셔너리

* 데이터 추가/수정 가능
* 시쿼스가 아니기 때문에 인덱싱 불가
  </br>→ key 값으로 value 값 받아와야 함
* for문이나 if문에서 딕셔너리 활용 시
  </br>→ .items() 또는 .keys() 또는 .values() 메소드 활용 (인수 없는 메소드)
  </br>→ 3가지 메소드 모드 리스트 타입으로 활용됨

```
## dict 함수 사용해 딕셔너리 생성 (3가지 방법)
## n차원 리스트
list1 = [ ['boy', '소년'], ['school', '학교'], ['book', '책'] ]
dic = dict(list1)
print(dic)

## 튜플로 구성된 리스트
list2 = [ ('boy', '소년'), ('school', '학교'), ('book', '책') ]
dic = dict(list2)
print(dic)

## 가변 kwarg (변수명이 key)
## dict() 사용 시 키값에 int 삽입 불가
list3 = dict(boy= '소년', school='학교', book='책')  
dic = dict(list3)
print(dic)
```

```
## .items() 에서 key, value 로 쉽게 가져올 수 있음
score_of_class = {
    "class1": (77, 85, 67, 80, 90),
    "class2": (60, 80, 75, 98, 82),
    "class3": (72, 77, 65, 83, 95),
    "class4": (92, 70, 85, 67, 88)
}

print("class3 :", score_of_class["class3"])
for key, value in score_of_class.items() :
    print("[", key, "반]")
    print("총점 :", sum(value))
    print("최고점 :", max(value))
    print("최저점 :", min(value))
```

##3. 지역변수&전역변수

1. 지역변수
  * 함수 안에서만 사용
2. 전역변수
  * 프로그램 전체에서 사용
  * 함수 안에서도 사용 가능
  * 만약 전역변수와 같은 이름의 변수에 변화(값 수정 등)를 준다면,
  </br>지역변수처럼 함수 내에서만 변화 적용되고 함수에서 나오면 다시 원래 전역변수 값으로 돌아옴
  </br>→ 전역변수 값 자체를 바꾸고 싶다면 *global* 예약어 사용해야 함
  </br>→ global은 (함수 실행 후) 기존 전역변수 id를 함수에서의 global에서 쓰인 주소로 바꿔줌...!

```
def test1(x):
  x = 3
  return x

a = 5
print(a)
test1(a)
print(a)   # 다시 원래 전역변수 값인 5로 돌아옴
print("---")


def test2():
  a = 3
  print("함수 내 변수 a의 id :", id(a))
  return a

print("전역변수 a의 id :", id(a))
print(test2())
print(a)   # 다시 원래 전역변수 값인 5로 돌아옴
print("전역변수 a의 id :", id(a))   # 함수 실행 전후 전역변수 id는 같음 (함수에서의 a는 다른 id 가짐)
```

```
# 함수를 통해 전역변수를 바꾸고 싶을 때
def test3():
  global a
  a = 3
  print("함수 내 global 변수 a의 id :", id(a))
  return a

a = 5
print("함수 실행 전, 전역변수 a의 id :", id(a))
print(a)
test3()
print(a)   # 전역변수 값이 바뀌었음
print("함수 실행 후, 전역변수 a의 id :", id(a))   
    # 함수 실행 전과 실행 후 전역변수의 id 바뀜. 즉, 전역변수의 함수 내에서 썼던 a의 id로 바뀜
```

```
### 함수명 뒤 소괄호 붙이면 '호출(call)'을,
### 생략하면 함수 자체를 불러오는 것을 의미
import types

def func() :
    print("hello")

a = func
print(a)
a()

b = func()
print(b)

############### 타입체크 #################
def callback(fn) :
    if type(fn) == types.FunctionType :
        fn()
    else :
        print(type(fn), end=": ")
        print("아규먼트로 함수를 주세요")

callback(a)
callback(b)
```

### 6-2. filter&map 함수

filter(*조건지정함수*, *대상리스트*)

* 리스트를 아규먼트로 받아 조건에 맞는 것만 골라냄
</br>→ 조건 지정에는 함수가 필요함
</br>→ 지금은 컴프리핸션으로 대체 가능
* 조건 함수의 리턴값은 무조건 bool 형태여야 함

```
phone = [119, 97674223, 1195524421]

li = list(filter(lambda x:x > 100000, phone))
print(li)

# 아래 제너레이터와 같음..
print([i for i in phone if i > 100000])
```

```
## 참고)
## filter 메소드와 generator 중 generater가 훨씬 빠름!!!!!!!
## 결론: filter 함수 버리쟈....★
%timeit list(filter(lambda x:x > 100000, phone))
%timeit [i for i in phone if i > 100000]
```

map(*변환함수*, *대상리스트*)
* 리스트를 아규먼트로 받아 전체 요소를 일괄 변환한 리스트 생성

```
phone = [119, 97674223, 1195524421]

## 예1
list1= list(map(str, phone))
print(list1)

## 예2
list2 = list(map(lambda x:len(str(x)), phone))
print(list2)
```

### 9-4. 이터러블(Iterable) 객체

* 객체에 속한 멤버들을 하나하나 꺼내쓸 수 있는 객체 (꺼내는 것을 '이터'라고 함)
</br>→ for 문에서 사용 가능함!</br>
ex. sequence, range(), dict ...
* 이터러블 객체는 \_\_iter\__ 와 \_\_next\__ 메서드를 갖고 있어야 함

```
class Seq:
    def __init__(self, data):
        self.data = data
        self.index = -2
    def __iter__(self):
        return self
    def __next__(self):
        self.index += 2
        if self.index >= len(self.data):
            raise StopIteration
        return self.data[self.index:self.index + 2]

solarterm = Seq("입춘우수경칩춘분청명곡우입하소만망종하지소서대서")
for k in solarterm:
    print(k, end = ',')
```

## 10. 데코레이터(@)

### 10-1. property

변수 = property(getter, setter)</br>
* r value로 쓰면(변수 변경 등) getter
* l value로 쓰면(반환, 출력 등) setter
* 함수를 파라미터로 하는 고차함수!

```
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def getAging(self):
    return self.age

  def setAging(self, year):
    self.age += year

  aging = property(getAging, setAging)   # getter와 setter를 aging으로 호출 가능

ir = Person('Irene', 25)
ir.aging = 2   # ir.getAging(2)와 같음
print(ir.aging)   # ir.setAging()와 같음
```

**데코레이터**
* getter 함수 윗줄에 → *@property*
* setter 함수 윗줄에 → *@메서드명.setter*
* 주의 : getter와 setter 함수의 이름이 같아야 함!

```
class Person:
  profile = []

  @classmethod
  def print(cls):
    print("=============================")
    print("name\tage\tjob\t\t")
    for i in cls.profile:
      print(str(i))    # 오버라이딩된 str 함수 사용
    print("=============================")

  def __init__(self, name, age, job):
    self.name = name
    self.age = age
    self.job = job
    Person.profile.append(self)

  def __str__(self):   # 오버라이딩
    return f"{self.name}\t{self.age}\t{self.job}"

Person.print()
print()

p1 = Person('Irene', 25, 'student')
p2 = Person('Tom', 33, 'developer')
Person.print()
print()

p3 = Person('John', 18, 'student')
p4 = Person('Lisa', 52, 'marketer')
Person.print()
print()
```
