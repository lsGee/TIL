# Pandas 심화

```
import pandas as pd
```

## Load Data

```
# order = pd.read_csv("https://bit.ly/dsa-01-order", index_col="id")
order = pd.read_csv("https://bit.ly/dsa-01-order")

order.set_index("id", inplace=True)
# order = order.set_index("id")

order.head()
```

## 정렬하기
pandas에서 데이터프레임 정렬 방법은 크게 아래 2가지다.
1. 인덱스 기준 정렬
2. 특정 컬럼 기준 정렬

### 1. 인덱스로 정렬하기

```
# id 내림차순
order.sort_index(ascending=False)
```

### 2. 컬럼으로 정렬하기

```
# price 높은순(내림차순)
order.sort_values(by="price", ascending=False)
```

```
# Series 에서 정렬 시 by 옵션 값은 필요 없음
order["price"].sort_values()
```

### 추가) 인덱스 정렬 응용

```
# 컬럼명을 오름차순으로 정렬
order.sort_index(axis="columns")
```

### 추가) 컬럼 정렬 응용

```
# 2개 이상의 컬럼 기준으로 각각 내림차순/오름차순으로 정렬
order.sort_values(by=["price", "date"], ascending=[False, True])
```

## Pivot Table
: 통계연산의 결과를 눈으로 확인하고 싶을 때 사용

* `NaN` 은 배제된다
* 디폴트는 평균값(`mean`)

```
pd.pivot_table(order, values="price", index="address")

## 아래와 같다
# order.pivot_table(values="price", index="address")
```

```
pd.pivot_table(order, values="price", index="address", aggfunc=["mean", "sum"])
```

```
pd.pivot_table(order, values="price", index=["date", "state"])
```

```
## Numpy 와 함께 쓸 수도 있다
import numpy as np

pd.pivot_table(order, values="price", index="address", aggfunc=[np.mean, np.sum])
```

```
## method chaining 가능가능
table = pd.pivot_table(order, values="price", index=["date", "state"])

table.sort_values(by="price", ascending=False).head(3)
```

## Group by
: 통계연산 후 결과를 다시 활용하고자 할 때 사용 (pivot_table보다 10배 정도 빠름)

* 또 다른 연산을 해야 할 때
* DB에 넣어야 할 때
* 머신러닝 알고리즘에 넣을 때
* 파일로 저장할 때

```
order.groupby("state")["price"].mean()
```

```
order.groupby(["date", "state"]).sum()
```

```
## method chaining 가능가능
order.groupby(["date", "state"]).sum().sort_values(by="date", ascending=False).head(5)
```

## apply

```
def is_vip(p):
  if p >= 500:
    return "VIP"
  else:
    return "Non_VIP"
  
order["price"].apply(is_vip)
```

```
## axis 옵션을 활용해 여러 컬럼에 대한 함수를 apply 에 사용할 수도 있다
def is_vip(row):
  if row["price"] >= 500 and row["state"] == "confirmed":
    return "VIP"
  else:
    return "non-VIP"

order.apply(is_vip, axis="columns")
```

## 여러 개의 데이터를 하나로 합치기

### 1. concat (= Concatenate)

#### 1-1. 위아래로 합치기

```
url = "https://bit.ly/dsa-02-order-201701"

order201701 = pd.read_csv(url, index_col="id").head()
order201701.head()
```

```
url = "https://bit.ly/dsa-02-order-201702"

order201702 = pd.read_csv(url, index_col="id").head()
order201702.head()
```

```
# 위아래로 합치기
order = pd.concat([order201701, order201702])
order
```

#### 1-2. 왼쪽, 오른쪽으로 합치기

```
url = "https://bit.ly/dsa-02-order-how"

how = pd.read_csv(url, index_col="id")
how
```

```
# 왼쪽, 오른쪽으로 합치기 (axis의 기본값은 index로, 위아래로 합친다는 의미)
# 인덱스를 기준으로 삼아서 합침
pd.concat([order, how], axis="columns")
```

### 2. merge

```
url = "https://bit.ly/dsa-02-user"

user = pd.read_csv(url, index_col="id")
user
```

```
url = "https://bit.ly/dsa-02-product"

product = pd.read_csv(url, index_col="id")
product
```

```
## 두 데이터프레임의 서로 다른 이름의 컬럼을 기준으로 삼기
# 컬럼 이름이 서로 같다면 on 옵션 하나만 사용하면 됨
pd.merge(order, user, left_on="user_id", right_on="id")
```

#### how 옵션
* `how="inner"` :  교집합
* `how="outer"` :  합집합
* `how="left"` :  왼쪽을 살리고 오른쪽 버림 (권장)
* `how="right"` :  오른쪽 살리고 왼쪽 버림

```
## outer
pd.merge(order, user, left_on = "user_id", right_on = "id", how="outer")
```

```
## left
pd.merge(order, user, left_on="user_id", right_on="id", how="left")
```
