## Data Load

```python
import pandas as pd
```

```python
train = pd.read_csv("../data/train.csv", index_col="PassengerId")
train.head()
```

```python
test = pd.read_csv("../data/test.csv", index_col="PassengerId")
test.head()
```

## Preprocessing

### Encode Sex

```python
train["Sex_encode"] = train["Sex"].replace("male", 0).replace("female", 1)
train[["Sex", "Sex_encode"]].head()
```

```python
test["Sex_encode"] = test["Sex"].replace("male", 0).replace("female", 1)
test[["Sex", "Sex_encode"]].head()
```

### Encode Embarked

```python
train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"

print(train.shape)
train[["Embarked", "Embarked_C", "Embarked_Q", "Embarked_S"]].head()
```

```python
test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"

print(test.shape)
test[["Embarked", "Embarked_C", "Embarked_Q", "Embarked_S"]].head()
```

### Age Type

```python
train_p1 = train.loc[train["Pclass"] == 1, "Age"].mean()
train_p2 = train.loc[train["Pclass"] == 2, "Age"].mean()
train_p3 = train.loc[train["Pclass"] == 3, "Age"].mean()
print(train_p1)
print(train_p2)
print(train_p3)
```

```python
train.loc[(train["Pclass"] == 1) & (train["Age"].isnull()), "Age"] = train_p1
train.loc[(train["Pclass"] == 2) & (train["Age"].isnull()), "Age"] = train_p2
train.loc[(train["Pclass"] == 3) & (train["Age"].isnull()), "Age"] = train_p3
train[train["Age"].isnull()]
```

```python
test_p1 = test.loc[test["Pclass"] == 1, "Age"].mean()
test_p2 = test.loc[test["Pclass"] == 2, "Age"].mean()
test_p3 = test.loc[test["Pclass"] == 3, "Age"].mean()
print(test_p1)
print(test_p2)
print(test_p3)
```

```python
test.loc[(test["Pclass"] == 1) & (test["Age"].isnull()), "Age"] = test_p1
test.loc[(test["Pclass"] == 2) & (test["Age"].isnull()), "Age"] = test_p2
test.loc[(test["Pclass"] == 3) & (test["Age"].isnull()), "Age"] = test_p3
test[test["Age"].isnull()]
```

```python
train.loc[train["Age"] < 15, "AgeType"] = "Young"
train.loc[(15 <= train["Age"]) & (train["Age"] < 30), "AgeType"] = "Medium"
train.loc[train["Age"] >= 30, "AgeType"] = "Old"

train["Age_Young"] = train["AgeType"] == "Yong"
train["Age_Medium"] = train["AgeType"] == "Medium"
train["Age_Old"] = train["AgeType"] == "Old"

print(train.shape)
train[["Age", "Age_Young", "Age_Medium", "Age_Old"]].tail()
```

```python
test.loc[test["Age"] < 15, "AgeType"] = "Young"
test.loc[(15 <= test["Age"]) & (test["Age"] < 30), "AgeType"] = "Medium"
test.loc[test["Age"] >= 30, "AgeType"] = "Old"

test["Age_Young"] = test["AgeType"] == "Yong"
test["Age_Medium"] = test["AgeType"] == "Medium"
test["Age_Old"] = test["AgeType"] == "Old"

print(test.shape)
test[["Age", "Age_Young", "Age_Medium", "Age_Old"]].tail(10)
```

###  Fill in missing Fare

```python
test["Fare"] = test["Fare"].fillna(0)   # 1개밖에 없어서 걍 0을 넣어준 것
test[test["Fare"].isnull()]
```

### Family Type

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
train.loc[train["FamilySize"] == 1, "FamilyType"] = "Single"
train.loc[train["FamilySize"] >= 2, "FamilyType"] = "Nuclear"
train.loc[train["FamilySize"] >= 5, "FamilyType"] = "Big"

train["F_Single"] = train["FamilyType"] == "Single"
train["F_Nuclear"] = train["FamilyType"] == "Nuclear"
train["F_Big"] = train["FamilyType"] == "Big"

train[["SibSp", "Parch","FamilySize", "F_Single", "F_Nuclear", "F_Big"]].head()
```

```python
test["FamilySize"] = test["SibSp"] + test["Parch"] + 1
test.loc[test["FamilySize"] == 1, "FamilyType"] = "Single"
test.loc[test["FamilySize"] >= 2, "FamilyType"] = "Nuclear"
test.loc[test["FamilySize"] >= 5, "FamilyType"] = "Big"

test["F_Single"] = test["FamilyType"] == "Single"
test["F_Nuclear"] = test["FamilyType"] == "Nuclear"
test["F_Big"] = test["FamilyType"] == "Big"

test[["SibSp", "Parch", "FamilySize", "F_Single", "F_Nuclear", "F_Big"]].head()
```

### Cabin

```python
## Cabin 정보의 유무
train.loc[train["Cabin"].isnull(), "isCabin"] = 0
train.loc[train["isCabin"].isnull(), "isCabin"] = 1

print(train["isCabin"].shape)
train[["Cabin", "isCabin"]].head()
```

```python
test.loc[test["Cabin"].isnull(), "isCabin"] = 0
test.loc[test["isCabin"].isnull(), "isCabin"] = 1

print(test["isCabin"].shape)
test[["Cabin", "isCabin"]].tail()
```

```python
## Cabin 수
train["Cabin_count"] = train["Cabin"].str.count(" ") + 1
train.loc[train["Cabin_count"].isnull(), "Cabin_count"] = 0

print(train["Cabin_count"].shape)
train["Cabin_count"].tail(20)
```

```python
## Cabin 수
test["Cabin_count"] = test["Cabin"].str.count(" ") + 1
test.loc[test["Cabin_count"].isnull(), "Cabin_count"] = 0

print(test["Cabin_count"].shape)
test["Cabin_count"].head(20)
```

## Train

```python
feature_names = ["Pclass", "Sex_encode", "Fare",
                 "Embarked_C", "Embarked_Q", "Embarked_S",
                 "Age_Young", "Age_Medium", "Age_Old",
                 "F_Single", "F_Nuclear", "F_Big",
                ]
feature_names
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
y_train = train["Survived"]
print(y_train.shape)
y_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

### Use Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth = 11)
model
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
print(prediction.shape)
prediction[0:5]
```

```python
submit = pd.read_csv("../data/gender_submission.csv", index_col="PassengerId")
print(submit.shape)

submit["Survived"] = prediction
print(submit.shape)
submit.head()
```

```python
submit.to_csv("../data/decision-tree.csv")
```
